```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df
```

```python
df['sex'] = df['sex'].replace({'Male': 'M', 'Female': 'F'})
df
```

```python
df['internet'] = df['internet'].fillna(df['internet'].mode()[0])
df
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.columns
```

```python
fig, axes = plt.subplots(1,2,figsize=(14,7))
axes[0].hist(df['absences'],bins = 30)
axes[0].set_title('Absences')
axes[0].set_xlabel('Frequency')
axes[0].set_ylabel('Count')
axes[1].plot(sorted(df['absences']),'o-')
axes[1].set_title('Sorted Absences')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Absences')
```

```python
plt.boxplot(df['G1'])
plt.title('G1')
plt.show()
```

```python
plt.boxplot(df['G2'])
plt.title('G2')
plt.show()
```

```python
q1_g2 = df['G2'].quantile(0.25)
q3_g2 = df['G2'].quantile(0.75)
iqr_g2 = q3_g2 - q1_g2
min_g2 = q1_g2 - 1.5 * iqr_g2
max_g2 = q3_g2 + 1.5 * iqr_g2
df['G2'] = np.where(df['G2'] < min_g2, min_g2, df['G2'])
df['G2'] = np.where(df['G2'] > max_g2, max_g2, df['G2'])
```

```python
plt.boxplot(df['G2'])
plt.title('G2')
plt.show()
```

```python
plt.boxplot(df['G3'])
plt.title('G3')
plt.show()
```

```python
q1_g3 = df['G3'].quantile(0.25)
q3_g3 = df['G3'].quantile(0.75)
iqr_g3 = q3_g3 - q1_g3
min_g3 = q1_g3 - 1.5 * iqr_g3
max_g3 = q3_g3 + 1.5 * iqr_g3
df['G3'] = np.where(df['G3'] < min_g3, min_g3, df['G3'])
df['G3'] = np.where(df['G3'] > max_g3, max_g3, df['G3'])
```

```python
plt.boxplot(df['G3'])
plt.title('G3')
plt.show()
```

```python
df.columns
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
axes[0].hist(df['age'], bins=15, color='blue')
axes[0].set_title('Age Distribution')
axes[1].hist(df['studytime'], bins=15, color='blue')
axes[1].set_title('Study Time Distribution')
axes[2].hist(df['absences'], bins=15, color='blue')
axes[2].set_title('Absences Distribution')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize = (18, 5))
axes[0].hist(df['G1'], bins=15, color='blue')
axes[0].set_title('G1 Distribution')
axes[1].hist(df['G2'], bins = 15, color = 'blue')
axes[1].set_title('G2 Distribution')
axes[2].hist(df['G3'], bins = 15, color = 'blue')
axes[2].set_title('G3 Distribution')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

df.groupby('sex')['G1'].mean().plot(kind='pie', autopct='%1.1f%%', ax=axes[0])
axes[0].set_xlabel('G1')
axes[0].set_ylabel('')
axes[0].set_title('G1 By Gender')
df.groupby('sex')['G2'].mean().plot(kind='pie', autopct='%1.1f%%', ax=axes[1])
axes[1].set_xlabel('G2')
axes[1].set_ylabel('')
axes[1].set_title('G2 By Gender')
df.groupby('sex')['G3'].mean().plot(kind='pie', autopct='%1.1f%%', ax=axes[2])
axes[2].set_xlabel('G3')
axes[2].set_ylabel('')
axes[2].set_title('G3 By Gender')
plt.show()
```
